```
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = [8, 8]
plt.rcParams.update({'font.size': 18})

x = 3 #true slope
a = np.arange(-2,2,0.25)
a = a.reshape(-1, 1) #making a column vector
b = x*a + np.random.randn(*a.shape) # Add noise

# True linear relationship
plt.plot(a, x*a, Color='k', LineWidth=2, label='True line') 
# Noisy measurement
plt.plot(a, b, 'x', Color='r', MarkerSize = 10, label='Noisy data')

#Economy SVD decomposition
U, S, VT = np.linalg.svd(a,full_matrices=False)
# Least-square fit: b = ax = USV^T x => x = V.S^(-1).U^T.b
xtilde = VT.T @ np.linalg.inv(np.diag(S)) @ U.T @ b 

plt.plot(a, xtilde * a,'--',Color='b',LineWidth=4, label='Regression line')

plt.xlabel('a')
plt.ylabel('b')

plt.grid(linestyle='--')
plt.legend()
plt.show()
```

```
# pseudo-inverse of computing regression
xtilde2 = np.linalg.pinv(a) @ b 

# True linear relationship
plt.plot(a, x*a, Color='k', LineWidth=2, label='True line') 
# Noisy measurement
plt.plot(a, b, 'x', Color='r', MarkerSize = 10, label='Noisy data')
plt.plot(a, xtilde2 * a,'--',Color='b',LineWidth=4, label='Regression line')

plt.xlabel('a')
plt.ylabel('b')

plt.grid(linestyle='--')
plt.legend()
plt.show()
```
