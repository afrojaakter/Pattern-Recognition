```
import matplotlib.pyplot as plt
import numpy as np
import os
plt.rcParams['figure.figsize'] = [8, 8]
plt.rcParams.update({'font.size': 18})

# Load dataset
A = np.loadtxt(os.path.join('/content/drive/MyDrive/Data/PatternRecognition/hald_ingredients.csv'),delimiter=',')
b = np.loadtxt(os.path.join('/content/drive/MyDrive/Data/PatternRecognition/hald_heat.csv'),delimiter=',')

# Solve Ax=b using SVD
U, S, VT = np.linalg.svd(A,full_matrices=0)
# x = V.S^(-1).U.b, here '.' is for matrix multiplication
x = VT.T @ np.linalg.inv(np.diag(S)) @ U.T @ b

plt.plot(b, Color='k', LineWidth=2, label='Heat Data') # True relationship
plt.plot(A@x, '-o', Color='r', LineWidth=1.5, MarkerSize=6, label='Regression')
plt.xlabel('Mixture')
plt.ylabel('Heat [Cal/gram]')
plt.legend()
plt.show()
```

```
# Alternative Methods using pseudo-inverse:
x1 = np.linalg.pinv(A)@b
plt.plot(b, Color='k', LineWidth=2, label='Heat Data') # True relationship
plt.plot(A@x1, '-o', Color='r', LineWidth=1.5, MarkerSize=6, label='Regression')
plt.xlabel('Mixture')
plt.ylabel('Heat [Cal/gram]')
plt.legend()
plt.show()
```
