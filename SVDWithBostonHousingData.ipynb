#using all data to train

```
import matplotlib.pyplot as plt
import numpy as np
import os


# Load dataset
H = np.loadtxt(os.path.join('/content/Housing.csv'))
b = H[:,-1] # housing values in $1000s
A = H[:,:-1] # other factors

# Pad with ones for nonzero offset
A = np.pad(A,[(0,0),(0,1)],mode='constant',constant_values=1)

# Solve Ax=b using SVD
# Note that the book uses the Matlab-specific "regress" command
U, S, VT = np.linalg.svd(A,full_matrices=0)
x = VT.T @ np.linalg.inv(np.diag(S)) @ U.T @ b

plt.rcParams['figure.figsize'] = [16, 8]
plt.rcParams.update({'font.size': 18})

fig = plt.figure()
ax1 = fig.add_subplot(121)

plt.plot(b, Color='gray', LineWidth=2, label='Housing Value') # True relationship
plt.plot(A@x, '-o', Color='skyblue', LineWidth=1.5, MarkerSize=6, label='Regression')
plt.xlabel('Neighborhood')
plt.ylabel('Median Home Value [$1k]')
plt.legend()

ax2 = fig.add_subplot(122)
sort_ind = np.argsort(H[:,-1])
plt.plot(b[sort_ind], Color='gray', LineWidth=2, label='Housing Value') # True relationship
plt.plot(A[sort_ind,:]@x, '-o', Color='skyblue', LineWidth=1.5, MarkerSize=6, label='Regression')
plt.xlabel('Neighborhood')
plt.legend()

plt.show()
```

```
A_mean = np.mean(A,axis=0)
A_mean = A_mean.reshape(-1, 1)

A2 = A - np.ones((A.shape[0],1)) @ A_mean.T

for j in range(A.shape[1]-1):
    A2std = np.std(A2[:,j])
    A2[:,j] = A2[:,j]/A2std
    
A2[:,-1] = np.ones(A.shape[0])

U, S, VT = np.linalg.svd(A2,full_matrices=0)
x = VT.T @ np.linalg.inv(np.diag(S)) @ U.T @ b
x_tick = range(len(x)-1)+np.ones(len(x)-1)
plt.bar(x_tick,x[:-1])
plt.xlabel('Attribute')
plt.ylabel('Significance')
plt.xticks(x_tick)
plt.show()
```

```
# Load dataset
H = np.loadtxt(os.path.join('/content/Housing.csv'))
b = H[:,-1] # housing values in $1000s
A = H[:,:-1] # other factors

# Pad with ones for nonzero offset
A = np.pad(A,[(0,0),(0,1)],mode='constant',constant_values=1)

n = 253
btrain = b[:n]
Atrain = A[:n]
btest = b[n:]
Atest = A[n:]

# Solve Ax=b using SVD
# Note that the book uses the Matlab-specific "regress" command
U, S, VT = np.linalg.svd(Atrain,full_matrices=0)
x = VT.T @ np.linalg.inv(np.diag(S)) @ U.T @ btrain

plt.rcParams['figure.figsize'] = [16, 8]
plt.rcParams.update({'font.size': 18})

fig = plt.figure()
ax1 = fig.add_subplot(121)

plt.plot(btrain, Color='gray', LineWidth=2, label='Housing Value') # True relationship
plt.plot(Atrain@x, '-o', Color='skyblue', LineWidth=1.5, MarkerSize=6, label='Regression')
plt.xlabel('Neighborhood')
plt.title('train data')
plt.ylabel('Median Home Value [$1k]')
plt.legend()

ax2 = fig.add_subplot(122)
plt.plot(btest, Color='gray', LineWidth=2, label='Housing Value') # True relationship
plt.plot(Atest@x, '-o', Color='skyblue', LineWidth=1.5, MarkerSize=6, label='Regression')
plt.xlabel('Neighborhood')
plt.ylabel('Median Home Value [$1k]')
plt.title('Test data')
plt.legend()
plt.show()
```

#using all half of the data to train but with shuffle which leads to the better train and test results

```
# Load dataset
H = np.loadtxt(os.path.join('/content/Housing.csv'))
b = H[:,-1] # housing values in $1000s
A = H[:,:-1] # other factors

# Pad with ones for nonzero offset
A = np.pad(A,[(0,0),(0,1)],mode='constant',constant_values=1)

n = 253
P = np.random.permutation(506)
A = A[P,:]
b = b[P]
btrain = b[:n]
Atrain = A[:n]
btest = b[n:]
Atest = A[n:]

# Solve Ax=b using SVD
# Note that the book uses the Matlab-specific "regress" command
U, S, VT = np.linalg.svd(Atrain,full_matrices=0)
x = VT.T @ np.linalg.inv(np.diag(S)) @ U.T @ btrain

plt.rcParams['figure.figsize'] = [16, 8]
plt.rcParams.update({'font.size': 18})

fig = plt.figure()
ax1 = fig.add_subplot(121)

plt.plot(btrain, Color='gray', LineWidth=2, label='Housing Value') # True relationship
plt.plot(Atrain@x, '-o', Color='skyblue', LineWidth=1.5, MarkerSize=6, label='Regression')
plt.xlabel('Neighborhood')
plt.title('train data')
plt.ylabel('Median Home Value [$1k]')
plt.legend()

ax2 = fig.add_subplot(122)
plt.plot(btest, Color='gray', LineWidth=2, label='Housing Value') # True relationship
plt.plot(Atest@x, '-o', Color='skyblue', LineWidth=1.5, MarkerSize=6, label='Regression')
plt.xlabel('Neighborhood')
plt.ylabel('Median Home Value [$1k]')
plt.title('Test data')
plt.legend()
plt.show()
```
